```python
from sympy import symbols, Matrix, Transpose
import sympy as sp

# Define symbols
t, x00, x01, x02, v00, v01, v02, x10, x11, x12, v10, v11, v12, x20, x21, x22, v20, v21, v22, x30, x31, x32, v30, v31, v32 = symbols('t x00 x01 x02 v00 v01 v02 x10 x11 x12 v10 v11 v12 x20 x21 x22 v20 v21 v22 x30 x31 x32 v30 v31 v32')
x0 = Matrix([x00, x01, x02])
x1 = Matrix([x10, x11, x12])
x2 = Matrix([x20, x21, x22])
x3 = Matrix([x30, x31, x32])
v0 = Matrix([v00, v01, v02])
v1 = Matrix([v10, v11, v12])
v2 = Matrix([v20, v21, v22])
v3 = Matrix([v30, v31, v32])
x0t = x0 + v0 * t
x1t = x1 + v1 * t
x2t = x2 + v2 * t
x3t = x3 + v3 * t
x30t = x3t - x0t
x20t = x2t - x0t
x10t = x1t - x0t
f = x20t.dot(x10t.cross(x30t))
sp.expand(f)
```
